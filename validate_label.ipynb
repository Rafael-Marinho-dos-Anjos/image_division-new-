```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from skimage import io
```

```python
with open(r"instruments\cabo\Label\YOLO\frame_0132.txt", "r") as label:
    label = label.read().split(" ")
x = float(label[1])
y = float(label[2])
w = float(label[3])
h = float(label[4])
```

```python
img = io.imread(r"\\storage01\Robotica\Dataset CME\ferramentas montadas\cabo\fotos\frame_0132.jpg")
plt.imshow(img)
```

```python
x = int(x * img.shape[1])
y = int(y * img.shape[0])
w = int(w * img.shape[1])
h = int(h * img.shape[0])
img = cv2.rectangle(img, (x, y), (x + w, y + h), [255,0,0], 15)
plt.imshow(img)
```
